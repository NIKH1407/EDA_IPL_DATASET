# IPL DATASET DATA ANALYSIS

```python
#import important libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#loading the ipl matches dataset

ipl  = pd.read_csv("matches.csv")

ipl
```

```python
ipl.head() #for first five records of the data set
```

```python
ipl.shape # for no. of rows and no of column
```

```python
ipl["player_of_match"].value_counts() #for getting frequency of most man of the match awards
```

```python
ipl["player_of_match"].value_counts()[0:10]
```

```python
mom = ipl["player_of_match"].value_counts()[0:5]
```

```python
mom
```

```python
l = list(ipl["player_of_match"].value_counts()[0:5].keys())
```

```python
l
```

```python
#making bar plot of top 5 players with most man of the match awards

plt.figure(figsize = (8,5))
plt.bar(l,mom,color = "g")
plt.title("Player of match")
plt.show()
```

```python
#getting frequency of result column 

ipl["result"].value_counts()
```

```python
#getting frequency of toss_winner column 

ipl["toss_winner"].value_counts()
```

```python
batting_first = ipl[ipl["win_by_runs"] !=0] #team won by batting first

batting_first
```

```python
batting_first.head()
```

```python
#making histogram plot because win by run is a numerical column so we use hist plot

plt.figure(figsize =(5,5) )
plt.hist(batting_first["win_by_runs"])
plt.title("Distribution of Runs")
plt.xlabel("Run")
plt.ylabel("Matches")
plt.show()
```

```python
batting_first["winner"].value_counts()
```

```python
bf =batting_first["winner"].value_counts()[0:5] # for top 5 teams
```

```python
bf
```

```python
l1=list(batting_first["winner"].value_counts()[0:5].keys())
```

```python
l1
```

```python
#making bar graph of winning teams

plt.figure(figsize = (6,6))
plt.bar(l1[0:3],bf[0:3],color =["blue","Yellow","orange"])
plt.title("TOP 3 Teams")
plt.show()
```

```python
#making pie chart for %distribution of winning Teams

plt.figure(figsize = (12,12))
plt.pie(list(batting_first["winner"].value_counts()),labels = list(batting_first["winner"].value_counts().keys()),autopct ="%0.1f%%")
plt.show()
```

```python
batting_second = ipl[ipl["win_by_wickets"] !=0] #team won by batting second

batting_second
```

```python
batting_second.head()
```

```python
#making histogram  for frequency of win wrt to number of wicket

plt.hist(batting_second["win_by_wickets"],bins =30)
plt.title("win wrt to number of wicket")
plt.xlabel("wicket")
plt.ylabel("matches")
plt.show()
```
